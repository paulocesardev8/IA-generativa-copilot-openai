```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np

uploaded = files.upload()

image_path= next(iter(uploaded))
img = Image.open(image_path)
img = np.array(img)

if img.shape[-1] == 4:
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1, 4, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Imagem em Escala de Cinza")

img_blur = cv2.GaussianBlur(img_gray, (5,5), 0)

plt.subplot(1, 4, 3)
plt.imshow(img_blur, cmap='gray')
plt.title("Imagem Suavizada (Gaussiano)")

sobel_x = cv2.Sobel(img_blur, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img_blur, cv2.CV_64F, 0, 1, ksize=3)
img_edges = cv2.magnitude(sobel_x, sobel_y)

plt.subplot(1, 4, 4)
plt.imshow  (img_edges, cmap='gray')
plt.title("Detecção de Bordas (Sobel)")

plt.show()
```
