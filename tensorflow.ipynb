```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# Dados de treinamento (X) e seus rotulos (y) para uma classificaГ§ГЈo binГЎria simples
X = np.random.rand(100, 1) # exemplo com 100 amostras e 1 recurso
y = (X > 0.5).astype(np.float32) # Rotulos binarios (0 ou 1)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(1,)), # Camada oculta com 10 neurГґnios e funГ§ГЈo de ativaГ§ГЈo ReLu
    keras.layers.Dense(1, activation='sigmoid') # Camada de saГ­da com 1 neurГґnio e funГ§ГЈo de ativaГ§ГЈo sigmoid (para classificaГ§ГЈo binГЎria)
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=50, batch_size=32)
```

```python
# Exemplo de previsГЈo com novo dado (0.7)
new_data = np.array([[0.7]])
prediction = model.predict(new_data)
print("PrevisГЈo:", prediction)
```
